```python
import numpy as np
import pandas as pd
```

```python
df_ids = pd.read_csv('external_ids.csv')
df_ids
```

```python
df_ids['tmdb_id'] = pd.to_numeric(df_ids['tmdb_id'], errors='coerce').astype('Int32')
df_ids
```

```python
df_titles = pd.read_csv('titles.csv')
df_titles
```

```python
df_join = pd.merge(df_titles, df_ids, how='inner', on='film')
df_join
```

```python
df_facts = pd.read_csv('static_facts.csv')
df_facts
```

```python
df_facts['year'] = np.floor(pd.to_numeric(df_facts['year'], errors='coerce')).astype('Int16')
df_facts['duration'] = np.floor(pd.to_numeric(df_facts['duration'], errors='coerce'))
df_facts
```

```python
df_join = pd.merge(df_join, df_facts, how='inner', on='film')
df_join
```
